```python
import nltk.data
```

```python
a='data.txt'
```

```python
dataset=nltk.data.load(a,format='text')
```

```python
dataset
```

# splitting into sentences

```python
import nltk
nltk.download('punkt')
```

```python
sentences=nltk.sent_tokenize(dataset)
```

```python
sentences
```

# splitting into words

```python
words=nltk.word_tokenize(dataset)
```

```python
words
```

# eliminating stop words

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtered=[]
for word in words:
    if word not in stop_words:
        filtered.append(word)
```

```python
filtered
```

# finding frequency of each word

```python
frequency = nltk.FreqDist(filtered)
```

```python
frequency
```

# stemming

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
after_stemming={}
for word in filtered:
    after_stemming[word]=porter_stemmer.stem(word)
```

```python
after_stemming
```

# parts of speech tagging

```python
nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(filtered)
```

# score of word

```python
frequency_words = nltk.FreqDist(words)
```

```python
frequency_words
```

# score of sentence

```python
frequency_sentences = nltk.FreqDist(sentences)
```

```python
frequency_sentences
```
