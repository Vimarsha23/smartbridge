```python
import pandas as pd
filename='https://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/telemonitoring/parkinsons_updrs.data'
df = pd.read_csv(filename)
```

# dropping the column motor_UPDRS

```python
df.drop('motor_UPDRS',axis=1)
```

# 3 array operations

```python
import numpy
def arithmetic_operations(column):
    arr=df[column].as_matrix()
    print("Total=",numpy.sum(arr))
    print("Mean=",numpy.sum(arr)/len(arr))
    print("Minimum=",numpy.amin(arr))
    print("Maximum=",numpy.amax(arr))

for j in list(df):
    print(j)
    arithmetic_operations(j)
    print('**************')
```
