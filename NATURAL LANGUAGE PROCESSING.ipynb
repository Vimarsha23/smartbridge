If we replace the word motorcar in (1a) by automobile, to get (1b), the meaning of the sentence stays pretty much the same: (1) a. Benz is credited with the invention of the motorcar. b. Since everything else in the sentence has remained unchanged, we can conclude that the words motorcar and automobile have the same meaning, i.e. they are synonyms.
